```python
import py7zr
import os

# Path to your .7z file
seven_z_file_path = 'Mining_DataSet/trainingData.7z'

# Directory where you want to extract the files
extract_to_directory = 'extracted_data/'

# Create the directory if it doesn't exist
os.makedirs(extract_to_directory, exist_ok=True)

# Extracting the .7z file
with py7zr.SevenZipFile(seven_z_file_path, mode='r') as archive:
    archive.extractall(path=extract_to_directory)

print("Extraction completed.")
```

```python
import pandas as pd

# List files in the extracted directory
extracted_files = os.listdir(extract_to_directory)
print("Extracted files:", extracted_files)

# Assuming the extracted file is a CSV
csv_file_path = os.path.join(extract_to_directory, extracted_files[0])

# Read the CSV file into a pandas DataFrame
df = pd.read_csv(csv_file_path)

# Display the first few rows of the DataFrame
df.head()
```

```python
# Path to the .txt file containing the column names
column_names_file_path = 'Mining_DataSet/column_names.txt'

# Read the column names from the .txt file
with open(column_names_file_path, 'r') as file:
    column_names = file.read().strip().split()

# If the column names are comma-separated, use split(',') instead
# column_names = file.read().strip().split(',')
```

```python
csv_file = 'trainingData5.csv'
csv_file_path = os.path.join(extract_to_directory, csv_file)
    
# Load the CSV file without headers
df = pd.read_csv(csv_file_path, header=None)
    
# Assign the column names
df.columns = column_names
    
# Save the updated DataFrame back to the same file or a new file
updated_csv_file_path = os.path.join(extract_to_directory, f'updated_{csv_file}')
df.to_csv(updated_csv_file_path, index=False)
    
print(f"Updated {csv_file} with column names.")
```

```python
csv_file = 'trainingData4.csv'
csv_file_path = os.path.join(extract_to_directory, csv_file)
    
# Load the CSV file without headers
df = pd.read_csv(csv_file_path, header=None)
    
# Assign the column names
df.columns = column_names
    
# Save the updated DataFrame back to the same file or a new file
updated_csv_file_path = os.path.join(extract_to_directory, f'updated_{csv_file}')
df.to_csv(updated_csv_file_path, index=False)
    
print(f"Updated {csv_file} with column names.")
```

```python
# Loop through each CSV file and add the column names
for csv_file in extracted_files:
    csv_file_path = os.path.join(extract_to_directory, csv_file)
    
    # Load the CSV file without headers
    df = pd.read_csv(csv_file_path, header=None)
    
    # Assign the column names
    df.columns = column_names
    
    # Save the updated DataFrame back to the same file or a new file
    updated_csv_file_path = os.path.join(extract_to_directory, f'updated_{csv_file}')
    df.to_csv(updated_csv_file_path, index=False)
    
    print(f"Updated {csv_file} with column names.")
```
