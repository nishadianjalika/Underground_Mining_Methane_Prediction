```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load the dataset
data = pd.read_csv('extracted_data/updated_trainingData1.csv')
```

```python
# Extract data for a specific sensor, e.g., AN311
sensor_name = 'AN311'
sensor_data = data.filter(like=sensor_name)
print(sensor_data)
```

```python
print(sensor_data.iloc[0])
```

```python


# Plot the time series for the first period (row 0)
plt.figure(figsize=(10, 6))
plt.plot(sensor_data.iloc[0], label=f'{sensor_name} Readings')
plt.title(f'{sensor_name} Sensor Data - First 10-minute Period')
plt.xlabel('Time (seconds)')
plt.ylabel('Sensor Value')
plt.legend()
plt.show()
```

```python
# Visualize the distribution
plt.figure(figsize=(10, 6))
sensor_data.iloc[0].plot(kind='hist', bins=10)
plt.title(f'{sensor_name} Sensor Data Distribution - First 10-minute Period')
plt.xlabel('Sensor Value')
plt.ylabel('Frequency')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import re

# Load the dataset
df = pd.read_csv('extracted_data/updated_trainingData1.csv')

# List of sensors to process (assuming they are named consistently)
sensors = ['AN311', 'AN422', 'AN423', 'TP1721', 'RH1722', 'BA1723', 
           'TP1711', 'RH1712', 'BA1713', 'MM252', 'MM261', 'MM262', 
           'MM263', 'MM264', 'MM256', 'MM211', 'CM861', 'CR863', 
           'CR863', 'P_864', 'TC862', 'WM868', 'AMP1_IR', 'AMP2_IR', 
           'DMP3_IR', 'DMP4_IR', 'AMP5_IR', 'F_SIDE', 'V']

# Plot time values for each sensor
for sensor in sensors:
    # Extract columns related to the sensor
    sensor_columns = df.filter(regex=f'^{sensor}_value_').columns
    
    # Extract the time values (integers) from the column names
    time_values = [int(re.findall(r'\d+', col)[-1]) for col in sensor_columns]
    
    # Plot the time values
    plt.figure(figsize=(10, 4))
    plt.plot(time_values, [1]*len(time_values), 'bo-', label=sensor)
    plt.title(f'Time Values for Sensor {sensor}')
    plt.xlabel('Time (seconds)')
    plt.yticks([])
    plt.xlim(min(time_values) - 5, max(time_values) + 5)
    plt.ylim(0.9, 1.1)
    plt.grid(True)
    plt.legend()
    plt.show()
```
